```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(" ", "_")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna("MS", inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas.head()
```

```python
vendas[['Preço', 'Preço_com_frete']].describe()
```

```python
preco_errado = vendas.query('Preço > Preço_com_frete')
```

```python
vendas_correto = vendas.query('Preço < Preço_com_frete')
```

```python
vendas_correto.describe()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index()
```

```python
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel('/content/cliente_varejo (1).xlsx')
```

```python
cliente.head()
```

```python
vendas_correto.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how= 'left', on='cliente_Log')
vendas_cliente.head()
```

```python
#agg canal renda é uma variavel que eu criei
agg_idcanal_renda = vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').reset_index()
agg_idcanal_renda.head()
```

```python
agg_bandeira_idade = vendas_cliente.groupby('bandeira')['idade'].agg('mean').reset_index()
agg_bandeira_idade.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7, 4))
plt.bar(agg_bandeira_idade['bandeira'], agg_bandeira_idade['idade'], color='green')
plt.ylabel('media idade')
plt.title('idade por bandeira')
```

```python
fig = plt.figure(figsize=(7, 4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='green')
plt.ylabel('média renda')
plt.title('renda por canal de venda')
```

```python
vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data.head()
```

```python
fig = plt.figure(figsize=(15,4))
plt.xlabel('data da venda')
plt.ylabel('quantidade de vendas')
x = vendas_por_data['Data']
y = vendas_por_data['idcompra']

plt.plot(x, y, color ="green", linewidth = '2')
plt.show()
```

```python
agg_dpto_preco = vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index()
agg_dpto_preco.head()
```

```python
fig = plt.figure(figsize=(15, 4))
plt.bar(agg_dpto_preco['Nome_Departamento'], agg_dpto_preco['Preço_com_frete'], color='green')
plt.ylabel('Preço_com_frete')
plt.title('preço medio por departamento')
```
